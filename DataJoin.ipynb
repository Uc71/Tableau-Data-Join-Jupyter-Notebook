```python
#Import Dependency
import pandas as pd
```

```python
#Read in CSVs as Dataframes
Jan_df=pd.read_csv("201801-citibike-tripdata.csv")
Feb_df=pd.read_csv("201802-citibike-tripdata.csv")
Mar_df=pd.read_csv("201803-citibike-tripdata.csv")
Apr_df=pd.read_csv("201804-citibike-tripdata.csv")
May_df=pd.read_csv("201805-citibike-tripdata.csv")
```

```python
#Represent each DF with a variable and concatenate the variables into 1 new DF
Jan=Jan_df
Feb=Feb_df
Mar=Mar_df
Apr=Apr_df
May=May_df
JanMay_df=pd.concat([Jan,Feb,Mar,Apr,May])
```

```python
#Eliminated start stations with Latitudes >45
Goodstarts_df=JanMay_df[JanMay_df["start station latitude"]<45]
```

```python
#Eliminated end stations with Latitudes >45
Goodstartends_df=Goodstarts_df[Goodstarts_df["end station latitude"]<45]
```

```python
#I Gave the columns more elegant and descriptive names
Goodstartends_df.rename(columns={"tripduration":"Trip Duration (s)","starttime":"Start Time (ISO)","stoptime":"Stop Time (ISO)","start station id":"Start Station ID","start station name":"Start Station Address","start station latitude":"Start Station Latitude","start station longitude":"Start Station Longitude","end station id":"End Station ID","end station name":"End Station Address","end station latitude":"End Station Latitude","end station longitude":"End Station Longitude","bikeid":"Bike ID","usertype":"User Type","birth year":"User Birth Year","gender":"Gender (0=Unspecified, 1=Male, 2=Female)"},inplace=True)
```
